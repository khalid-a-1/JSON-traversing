```python
def postorderTreeJSON(d):
    #print(d)
    if type(d) == str:
        print(d)
    elif type(d)== list:
        for obj in d:
            postorderTreeJSON(obj)
        print(d)
    elif type(d) == dict:
        for k, v in d.items():
            postorderTreeJSON(v)
            print(k)
```

```python
postorderTreeJSON(root1)
```
